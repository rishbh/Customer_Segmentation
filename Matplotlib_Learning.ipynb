#Plot a bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,60,80,20]

#convert labels to xpos numeric value
xpos=np.arange(len(company) )

plt.xticks(xpos,company)#convert all numeric values at xposition to company name
#plotting a barchart
plt.bar( xpos-0.1, revenue,width=0.2,label='Revenue')
plt.bar(xpos+0.1 , profit,width=0.2 ,label='Profit')
plt.title("Revenue and Profit of company ")
plt.xlabel('Company names')
plt.ylabel('Revenue(in billions)')
plt.legend()
plt.show()
```

#Plot a histogram


```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist( [blood_sugar_men ,blood_sugar_women], bins=[80,100,120,] , rwidth=0.95 , color=['green','orange'],label=['Men' , 'Women']  )
plt.xlabel('Blood Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Analysis')
plt.legend()
plt.show()
```

# Pie Chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.axis('equal')
plt.pie(exp_vals , labels=exp_labels  ,radius=1.5,explode=[0,0,0.40,0,0] ,autopct="%0.2f%%",startangle=45)
plt.title("Budget",color='black')
plt.show()
```
